# Proceso de QA

## Segunda parte: Evaluación de respuestas

### Se usará los resultados generados en: QA--Generación de respuestas-codigo_etica: codigo_etica_ix_baufest-etica-2_v3.json

El objetivo de esta sección es evaluar las respuestas generadas por distintos prompts.

__Requisitos:__ Importamos las siguientes librerías

```python
import sys
sys.path.append('../lib')

import qagenevaluation as qaeval
import qagraphutils as qagraph
```

__Paso 1:__ Seleccionamos algunos archivos para evaluar

En nuestro caso seleccionaremos: _codigo_etica_v3.json_ que contiene respuestas generadas:
- prompt simple: simple_prompt
- prompt usado en demo3: demo3_prompt
- prompt experimental: experimental_prompt

```python
filename_to_eval = "codigo_etica_ix_baufest-etica-2_v3.json"
```

__Paso 2:__ Procedemos a genera la evaluación usando solamente Azure Open AI para mayores detalles revisar qaoaievalutils

Notar que estaremos generando las siguientes evaluaciones:
- codigo_etica_v3_simple_vs_demo3
- codigo_etica_v3_simple_vs_experimental
- codigo_etica_v3_demo3_vs_experimental

```python
eval_result=qaeval.evaluate_responses(filename_to_eval=filename_to_eval,
                                   response_a_field="simple_prompt",
                                   response_b_field="demo3_prompt", 
                                   filename_result="codigo_etica_v3_simple_vs_demo3.json")
qagraph.show_evaluate_responses_as_bar_graph(filename_to_eval=eval_result)
qagraph.show_evaluate_responses_as_box_graph(filename_to_eval=eval_result)
```

```python
eval_result=qaeval.evaluate_responses(filename_to_eval=filename_to_eval,
                                   response_a_field="simple_prompt",
                                   response_b_field="experimental_prompt", 
                                   filename_result="codigo_etica_v3_simple_vs_experimental.json")
qagraph.show_evaluate_responses_as_bar_graph(filename_to_eval=eval_result)
qagraph.show_evaluate_responses_as_box_graph(filename_to_eval=eval_result)
```

```python
eval_result=qaeval.evaluate_responses(filename_to_eval=filename_to_eval,
                                   response_a_field="demo3_prompt",
                                   response_b_field="experimental_prompt", 
                                   filename_result="codigo_etica_v3_demo3_vs_experimental.json")
qagraph.show_evaluate_responses_as_bar_graph(filename_to_eval=eval_result)
qagraph.show_evaluate_responses_as_box_graph(filename_to_eval=eval_result)
```
