```python
import re
import json

debug = False

archivo_path = "indice.txt"
industrias = ["Agroindustria", "Alimentos / Bebidas", "Construcción", "Utilities", "Finanzas", "Gobierno", "Manufactura", "Medios", "Retail", "Salud / Laboratorios", "Seguros", "Servicios", "Telecomunicaciones", "Educación"]

# Leer el archivo de texto
with open(archivo_path, 'r', encoding='utf-8') as file:
    texto = file.read()

# Inicializar el diccionario para almacenar las particiones por industria
industrias_lower = [industria.lower() for industria in industrias]

particiones = {industria: [] for industria in industrias_lower}
print(particiones)

# Dividir el texto en líneas
lineas = texto.split('\n')
if debug: print(len(lineas))

# Asignar cada línea a la industria correspondiente
particion = ""
for linea in lineas:
    lineaToRead = linea.strip().split()[0].lower()
    if debug: print(f"lineaToRead:{lineaToRead}")
    if lineaToRead in [elemento.lower() for elemento in industrias]:
        particion = lineaToRead
        if debug: print(f"Particionando por {particion}")
    else:
        if particion:
            if debug: print(f"Agregando {linea}")
            particiones[particion].append(linea.strip())

# Guardar cada partición en un archivo separado
for industria, particion in particiones.items():
    if debug: print(f"industria {industria} particion {particion}")
    with open(f"{industria.replace(' / ', '_')}_particion.txt", 'w', encoding='utf-8') as output_file:
        output_file.write(f"{industria}\n")
        for linea in particion:
            output_file.write(f"{linea}\n")      

result = json.dumps(particiones)
if debug: print(result)
    
print(particiones['retail'])
```
